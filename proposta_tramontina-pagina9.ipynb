```python
#Leitura do pdf
lista9_cotacao_02 = tabula.read_pdf(r"C:\Users\oneen\OneDrive\Área de Trabalho\FERRAMENTAS _ONE\cdp19023051109003634960000000.pdf", pages="9")
pagina9_cotacao_02 = lista9_cotacao_02[0]
pagina9_cotacao_02.columns = ['indice01','indice02','indice03','indice04','indice05','indice06','indice07','indice08','indice09','indice10']

#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material9 = pagina9_cotacao_02[pagina9_cotacao_02['indice01'].str.contains('Descrição\r')]
descricao_material9 = descricao_material9['indice01'].str.replace('Descrição\r','')
descricao_material9 = descricao_material9.reset_index(drop=True)


#Preço_Unitário

preco_unitario9 = pagina9_cotacao_02[pagina9_cotacao_02['indice04'].str.len() > 0]
preco_unitario9 = pagina9_cotacao_02['indice04'].str.contains('Preço Unitário sem I\r')
preco_unitario9 = pagina9_cotacao_02['indice04'].str.replace('Preço Unitário sem I\r','')
preco_unitario9 = preco_unitario9[preco_unitario9.str.len() > 0]
preco_unitario9 = preco_unitario9[~preco_unitario9.str.contains('Emb.\r')]
preco_unitario9 = preco_unitario9.reset_index(drop=True)
preco_unitario9 = preco_unitario9.apply(lambda x: str(x).replace(",","."))
preco_unitario9 = preco_unitario9.astype('float64')


#Preço total - fitragem de preço total

preco_total9 = pagina9_cotacao_02[pagina9_cotacao_02['indice05'].str.len() > 0]
preco_total9 = pagina9_cotacao_02['indice05'].str.contains('I Preço Total\r')
preco_total9 = pagina9_cotacao_02['indice05'].str.replace('I Preço Total\r','')
preco_total9 = preco_total9[preco_total9.str.len() > 0]
preco_total9 = preco_total9[~preco_total9.str.contains('Qtde.\r')]
preco_total9 = preco_total9.reset_index(drop=True)
preco_total9 = preco_total9.apply(lambda x: str(x).replace(".",""))
preco_total9 = preco_total9.apply(lambda x: str(x).replace(",","."))
preco_total9 = preco_total9.astype('float64')

#mesclando os dados atrávés de cada indice formando uma tabela

df_mesclar09= descricao_material9.to_frame().merge(preco_unitario9,left_index=True,right_index=True)
df_mesclar09= descricao_material9.to_frame().merge(preco_unitario9,left_index=True,right_index=True)
Tabela_09= df_mesclar09.merge(preco_total9,left_index=True,right_index=True)
Tabela_09['indice02'] = Tabela_09['indice05'] / Tabela_09['indice04']


#display
#display(descricao_material9)
#display(quantidade_1)
#display(preco_unitario9)
#display(preco_total9)
display(Tabela_09)
#display(cotacao_Tabela_09)
```
