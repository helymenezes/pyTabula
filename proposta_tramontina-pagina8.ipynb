```python
import tabula
import pandas as pd
from IPython.core.display import display_html
import re
```

```python
#Leitura do pdf
lista8_cotacao_02 = tabula.read_pdf(r"C:\Users\oneen\OneDrive\Área de Trabalho\FERRAMENTAS _ONE\cdp19023051109003634960000000.pdf", pages="8")
pagina8_cotacao_02 = lista8_cotacao_02[0]
pagina8_cotacao_02.columns = ['indice01','indice02','indice03','indice04','indice05','indice06','indice07','indice08','indice09','indice10']

#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material8 = pagina8_cotacao_02[pagina8_cotacao_02['indice01'].str.contains('Descrição\r')]
descricao_material8 = descricao_material8['indice01'].str.replace('Descrição\r','')
descricao_material8 = descricao_material8.reset_index(drop=True)


#Preço_Unitário - FIltragem de preço unitário da tabela1
preco_unitario8 = pagina8_cotacao_02[pagina8_cotacao_02['indice04'].str.len() > 0]
preco_unitario8 = pagina8_cotacao_02['indice04'].str.contains('Preço Unitário sem I\r')
preco_unitario8 = pagina8_cotacao_02['indice04'].str.replace('Preço Unitário sem I\r','')
preco_unitario8 = preco_unitario8[preco_unitario8.str.len() > 0]
preco_unitario8 = preco_unitario8[~preco_unitario8.str.contains('Emb.\r')]
preco_unitario8 = preco_unitario8.reset_index(drop=True)
preco_unitario8 = preco_unitario8.apply(lambda x: str(x).replace(",","."))
preco_unitario8 = preco_unitario8.astype('float64')

#Preço total - fitragem de preço total

preco_total8 = pagina8_cotacao_02[pagina8_cotacao_02['indice05'].str.len() > 0]
preco_total8 = pagina8_cotacao_02['indice05'].str.contains('I Preço Total\r')
preco_total8 = pagina8_cotacao_02['indice05'].str.replace('I Preço Total\r','')
preco_total8 = preco_total8[preco_total8.str.len() > 0]
preco_total8 = preco_total8[~preco_total8.str.contains('Qtde.\r')]
preco_total8 = preco_total8.reset_index(drop=True)
preco_total8 = preco_total8.apply(lambda x: str(x).replace(".",""))
preco_total8 = preco_total8.apply(lambda x: str(x).replace(",","."))
preco_total8 = preco_total8.astype('float64')

#mesclando os dados atrávés de cada indice formando uma tabela

df_mesclar08= descricao_material8.to_frame().merge(preco_unitario8,left_index=True,right_index=True)
df_mesclar08= descricao_material8.to_frame().merge(preco_unitario8,left_index=True,right_index=True)
df_Tabela_08= df_mesclar08.merge(preco_total8,left_index=True,right_index=True)
df_Tabela_08['indice02'] = df_Tabela_08['indice05'] / df_Tabela_08['indice04']


#display
#display(descricao_material8)
#display(quantidade_1)
#display(preco_unitario8)
#display(preco_total8)
display(df_Tabela_08)
#display(cotacao_df_Tabela_08)
```
