```python
#Leitura do pdf
lista2_cotacao_03 = tabula.read_pdf(r"Proposta One 01100523.pdf", pages="all")
pagina2_cotacao_03 = lista2_cotacao_03[2]
pagina2_cotacao_03.columns = ['indice01','indice02','indice03','indice04','indice05']
#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material2 = pagina2_cotacao_03.dropna()
descricao_material2 = descricao_material2.reset_index(drop=True)
descricao_material2 = descricao_material2['indice01']
#Preço_Unitário - FIltragem quantidade da tabela1
quantiade2 = pagina2_cotacao_03['indice04'].dropna()
quantiade2 = quantiade2.reset_index(drop=True)
quantiade2 = quantiade2.astype('float64')
#Preço total - fitragem de preço total
preco_total2 = pagina2_cotacao_03['indice05'].dropna()
preco_total2 = preco_total2.reset_index(drop=True)
preco_total2 = preco_total2.apply(lambda x: str(x).replace("R$ ",""))
preco_total2 = preco_total2.apply(lambda x: str(x).replace(".",""))
preco_total2 = preco_total2.apply(lambda x: str(x).replace(",","."))
preco_total2 = preco_total2.astype('float64')
#mesclando os dados atrávés de cada indice formando uma tabela
df_mesclar02= descricao_material2.to_frame().merge(quantiade2,left_index=True,right_index=True)
Tabela_02= df_mesclar02.merge(preco_total2,left_index=True,right_index=True)
Tabela_02['indice06'] = Tabela_02['indice05'] / Tabela_02['indice04']

display(Tabela_02)

```
