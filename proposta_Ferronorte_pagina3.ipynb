```python
#Leitura do pdf
lista3_cotacao_03 = tabula.read_pdf(r"Proposta One 01100523.pdf", pages="all")
pagina3_cotacao_03 = lista3_cotacao_03[3]
pagina3_cotacao_03.columns = ['indice01','indice02','indice03','indice04','indice05']
#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material3 = pagina3_cotacao_03.dropna()
descricao_material3 = descricao_material3.reset_index(drop=True)
descricao_material3 = descricao_material3['indice01']
#Preço_Unitário - FIltragem quantidade da tabela1
quantiade3 = pagina3_cotacao_03['indice04'].dropna()
quantiade3 = quantiade3.reset_index(drop=True)
quantiade3 = quantiade3.astype('float64')
#Preço total - fitragem de preço total
preco_total3 = pagina3_cotacao_03['indice05'].dropna()
preco_total3 = preco_total3.reset_index(drop=True)
preco_total3 = preco_total3.apply(lambda x: str(x).replace("R$ ",""))
preco_total3 = preco_total3.apply(lambda x: str(x).replace(".",""))
preco_total3 = preco_total3.apply(lambda x: str(x).replace(",","."))
preco_total3 = preco_total3.astype('float64')
#mesclando os dados atrávés de cada indice formando uma tabela
prdf_mesclar03= descricao_material3.to_frame().merge(quantiade3,left_index=True,right_index=True)
Tabela_03= prdf_mesclar03.merge(preco_total3,left_index=True,right_index=True)
Tabela_03['indice06'] = Tabela_03['indice05'] / Tabela_03['indice04']

display(Tabela_03)
```
