```python
#Leitura do pdf
lista4_cotacao_03 = tabula.read_pdf(r"Proposta One 01100523.pdf", pages="all")
pagina4_cotacao_03 = lista4_cotacao_03[4]
pagina4_cotacao_03.columns = ['indice01','indice02','indice03','indice04','indice05']

#Descrição material - Filtragem de descirção somente da tabela 01

#descricao_material4 = pagina4_cotacao_03.drop('indice01',axis=1)
descricao_material4 = pagina4_cotacao_03.dropna()
descricao_material4 = descricao_material4.reset_index(drop=True)
descricao_material4 = descricao_material4['indice01']

#Preço_Unitário - FIltragem quantidade da tabela1

quantiade4 = pagina4_cotacao_03['indice04'].dropna()
quantiade4 = quantiade4.reset_index(drop=True)
quantiade4 = quantiade4.astype('float64')

#Preço total - fitragem de preço total

preco_total4 = pagina4_cotacao_03['indice05'].dropna()
preco_total4 = preco_total4.reset_index(drop=True)
preco_total4 = preco_total4.apply(lambda x: str(x).replace("R$ ",""))
preco_total4 = preco_total4.apply(lambda x: str(x).replace(".",""))
preco_total4 = preco_total4.apply(lambda x: str(x).replace(",","."))
preco_total4 = preco_total4.astype('float64')

#mesclando os dados atrávés de cada indice formando uma tabela

df_mesclar04= descricao_material4.to_frame().merge(quantiade4,left_index=True,right_index=True)
Tabela_04= df_mesclar04.merge(preco_total4,left_index=True,right_index=True)
Tabela_04['indice06'] = Tabela_04['indice05'] / Tabela_04['indice04']

display(Tabela_04)

```
