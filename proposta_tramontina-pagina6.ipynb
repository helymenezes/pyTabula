```python
#Leitura do pdf
lista6_cotacao_02 = tabula.read_pdf(r"C:\Users\oneen\OneDrive\Área de Trabalho\FERRAMENTAS _ONE\cdp19023051109003634960000000.pdf", pages="6")
pagina6_cotacao_02 = lista6_cotacao_02[0]
pagina6_cotacao_02.columns = ['indice01','indice02','indice03','indice04','indice05','indice06','indice07','indice08','indice09','indice10']

#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material6 = pagina6_cotacao_02[pagina6_cotacao_02['indice01'].str.contains('Descrição\r')]
descricao_material6 = descricao_material6['indice01'].str.replace('Descrição\r','')
descricao_material6 = descricao_material6.reset_index(drop=True)


#Preço_Unitário - FIltragem de preço unitário da tabela1
preco_unitario6 = pagina6_cotacao_02[pagina6_cotacao_02['indice04'].str.len() > 0]
preco_unitario6 = pagina6_cotacao_02['indice04'].str.contains('Preço Unitário sem I\r')
preco_unitario6 = pagina6_cotacao_02['indice04'].str.replace('Preço Unitário sem I\r','')
preco_unitario6 = preco_unitario6[preco_unitario6.str.len() > 0]
preco_unitario6 = preco_unitario6[~preco_unitario6.str.contains('Emb.\r')]
preco_unitario6 = preco_unitario6.reset_index(drop=True)
preco_unitario6 = preco_unitario6.apply(lambda x: str(x).replace(".",""))
preco_unitario6 = preco_unitario6.apply(lambda x: str(x).replace(",","."))
preco_unitario6 = preco_unitario6.astype('float64')

#Preço total - fitragem de preço total

preco_total6 = pagina6_cotacao_02[pagina6_cotacao_02['indice05'].str.len() > 0]
preco_total6 = pagina6_cotacao_02['indice05'].str.contains('I Preço Total\r')
preco_total6 = pagina6_cotacao_02['indice05'].str.replace('I Preço Total\r','')
preco_total6 = preco_total6[preco_total6.str.len() > 0]
preco_total6 = preco_total6[~preco_total6.str.contains('Qtde.\r')]
preco_total6 = preco_total6.reset_index(drop=True)
preco_total6 = preco_total6.apply(lambda x: str(x).replace(".",""))
preco_total6 = preco_total6.apply(lambda x: str(x).replace(",","."))
preco_total6 = preco_total6.astype('float64')

#mesclando os dados atrávés de cada indice formando uma tabela


df_mesclar06= descricao_material6.to_frame().merge(preco_unitario6,left_index=True,right_index=True)
df_mesclar06= descricao_material6.to_frame().merge(preco_unitario6,left_index=True,right_index=True)
Tabela_06= df_mesclar06.merge(preco_total6,left_index=True,right_index=True)
Tabela_06['indice02'] = Tabela_06['indice05'] / Tabela_06['indice04']



#display(descricao_material6)
#display(preco_unitario6)
#display(preco_total6)
display(Tabela_06)
#display(cotacao_Tabela_06)
```
