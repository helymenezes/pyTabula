```python
#Leitura do pdf
lista0_cotacao_03 = tabula.read_pdf(r"Proposta One 01100523.pdf", pages="all")
pagina0_cotacao_03 = lista0_cotacao_03[0]
pagina0_cotacao_03.columns = ['indice01','indice02','indice03','indice04','indice05','indice06']
#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material1 = pagina0_cotacao_03.drop('indice01',axis=1)
descricao_material1 = descricao_material1.dropna()
descricao_material1 = descricao_material1.reset_index(drop=True)
descricao_material1 = descricao_material1['indice02']
#Preço_Unitário - FIltragem quantidade da tabela1
quantiade0 = pagina0_cotacao_03['indice05'].dropna()
quantiade0 = quantiade0.reset_index(drop=True)
quantiade0 = quantiade0.astype('float64')
#Preço total - fitragem de preço total
preco_total0 = pagina0_cotacao_03['indice06'].dropna()
preco_total0 = preco_total0.reset_index(drop=True)
preco_total0 = preco_total0.apply(lambda x: str(x).replace("R$ ",""))
preco_total0 = preco_total0.apply(lambda x: str(x).replace(".",""))
preco_total0 = preco_total0.apply(lambda x: str(x).replace(",","."))
preco_total0 = preco_total0.astype('float64')
#mesclando os dados atrávés de cada indice formando uma tabela
df_mesclar0= descricao_material1.to_frame().merge(quantiade0,left_index=True,right_index=True)
Tabela_0= df_mesclar0.merge(preco_total0,left_index=True,right_index=True)
Tabela_0['indice07'] = Tabela_0['indice06'] / Tabela_0['indice05']

display(Tabela_0)

```
