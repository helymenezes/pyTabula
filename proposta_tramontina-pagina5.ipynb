```python
import tabula
import pandas as pd
from IPython.core.display import display_html
import re
```

```python
#Leitura do pdf
lista5_cotacao_02 = tabula.read_pdf(r"C:\Users\oneen\OneDrive\Área de Trabalho\FERRAMENTAS _ONE\cdp19023051109003634960000000.pdf", pages="5")
pagina5_cotacao_02 = lista5_cotacao_02[0]
pagina5_cotacao_02.columns = ['indice01','indice02','indice03','indice04','indice05','indice06','indice07','indice08','indice09','indice10']

#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material5 = pagina5_cotacao_02[pagina5_cotacao_02['indice01'].str.contains('Descrição\r')]
descricao_material5 = descricao_material5['indice01'].str.replace('Descrição\r','')
descricao_material5 = descricao_material5.reset_index(drop=True)


#Preço_Unitário - FIltragem de preço unitário da tabela1
preco_unitario5 = pagina5_cotacao_02[pagina5_cotacao_02['indice04'].str.len() > 0]
preco_unitario5 = pagina5_cotacao_02['indice04'].str.contains('Preço Unitário sem I\r')
preco_unitario5 = pagina5_cotacao_02['indice04'].str.replace('Preço Unitário sem I\r','')
preco_unitario5 = preco_unitario5[preco_unitario5.str.len() > 0]
preco_unitario5 = preco_unitario5[~preco_unitario5.str.contains('Emb.\r')]
preco_unitario5 = preco_unitario5.reset_index(drop=True)
preco_unitario5 = preco_unitario5.apply(lambda x: str(x).replace(",","."))
preco_unitario5 = preco_unitario5.astype('float64')

#Preço total - fitragem de preço total

preco_total5 = pagina5_cotacao_02[pagina5_cotacao_02['indice05'].str.len() > 0]
preco_total5 = pagina5_cotacao_02['indice05'].str.contains('I Preço Total\r')
preco_total5 = pagina5_cotacao_02['indice05'].str.replace('I Preço Total\r','')
preco_total5 = preco_total5[preco_total5.str.len() > 0]
preco_total5 = preco_total5[~preco_total5.str.contains('Qtde.\r')]
preco_total5 = preco_total5.reset_index(drop=True)
preco_total5 = preco_total5.apply(lambda x: str(x).replace(".",""))
preco_total5 = preco_total5.apply(lambda x: str(x).replace(",","."))
preco_total5 = preco_total5.astype('float64')

#mesclando os dados atrávés de cada indice formando uma tabela

df_mesclar05= descricao_material5.to_frame().merge(preco_unitario5,left_index=True,right_index=True)
df_mesclar05= descricao_material5.to_frame().merge(preco_unitario5,left_index=True,right_index=True)
df_mesclar05= df_mesclar05.merge(preco_total5,left_index=True,right_index=True)
df_mesclar05['indice02'] = df_mesclar05['indice05'] / df_mesclar05['indice04']


#display
#display(descricao_material5)
#display(quantidade_1)
#display(preco_unitario5)
#display(preco_total5)
display(df_mesclar05)
#display(cotacao_df_mesclar05)
```
