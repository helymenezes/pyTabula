```python
#Leitura do pdf
lista1_cotacao_03 = tabula.read_pdf(r"Proposta One 01100523.pdf", pages="all")
pagina1_cotacao_03  = lista1_cotacao_03[1]
pagina1_cotacao_03.columns = ['indice01','indice02','indice03','indice04','indice05']
#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material1 = pagina1_cotacao_03.dropna()
descricao_material1 = descricao_material1.reset_index(drop=True)
descricao_material1 = descricao_material1['indice01']
#Preço_Unitário - FIltragem quantidade da tabela1
quantiade1 = pagina1_cotacao_03['indice04'].dropna()
quantiade1 = quantiade1.reset_index(drop=True)
quantiade1 = quantiade1.astype('float64')
#Preço total - fitragem de preço total
preco_total1 = pagina1_cotacao_03['indice05'].dropna()
preco_total1 = preco_total1.reset_index(drop=True)
preco_total1 = preco_total1.apply(lambda x: str(x).replace("R$ ",""))
preco_total1 = preco_total1.apply(lambda x: str(x).replace(".",""))
preco_total1 = preco_total1.apply(lambda x: str(x).replace(",","."))
preco_total1 = preco_total1.astype('float64')
#mesclando os dados atrávés de cada indice formando uma tabela
df_mesclar01= descricao_material1.to_frame().merge(quantiade1,left_index=True,right_index=True)
Tabela_01= df_mesclar01.merge(preco_total1,left_index=True,right_index=True)
Tabela_01['indice06'] = Tabela_01['indice05'] / Tabela_01['indice04']

display(Tabela_01)


```
