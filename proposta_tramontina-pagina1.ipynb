```python
#Leitura do pdf
lista1_cotacao_02 = tabula.read_pdf(r"C:\Users\oneen\OneDrive\Área de Trabalho\FERRAMENTAS _ONE\cdp19023051109003634960000000.pdf", pages="1")
pagina1_cotacao_02 = lista1_cotacao_02[0]
pagina1_cotacao_02.columns = ['indice01','indice02','indice03','indice04','indice05','indice06','indice07','indice08','indice09','indice10']

#Descrição material - Filtragem de descirção somente da tabela 01
descricao_material1 = pagina1_cotacao_02[pagina1_cotacao_02['indice01'].str.contains('Descrição\r')]
descricao_material1 = descricao_material1['indice01'].str.replace('Descrição\r','')
descricao_material1 = descricao_material1.reset_index(drop=True)


#Preço_Unitário - FIltragem de preço unitário da tabela1
preco_unitario1 = pagina1_cotacao_02[pagina1_cotacao_02['indice04'].str.len() > 0]
preco_unitario1 = pagina1_cotacao_02['indice04'].str.contains('Preço Unitário sem I\r')
preco_unitario1 = pagina1_cotacao_02['indice04'].str.replace('Preço Unitário sem I\r','')
preco_unitario1 = preco_unitario1[preco_unitario1.str.len() > 0]
preco_unitario1 = preco_unitario1[~preco_unitario1.str.contains('Emb.\r1')]
preco_unitario1 = preco_unitario1.reset_index(drop=True)
preco_unitario1 = preco_unitario1.apply(lambda x: str(x).replace(",","."))
preco_unitario1 = preco_unitario1.astype('float64')


#Preço total - fitragem de preço total

preco_total1 = pagina1_cotacao_02[pagina1_cotacao_02['indice05'].str.len() > 0]
preco_total1 = pagina1_cotacao_02['indice05'].str.contains('I Preço Total\r')
preco_total1 = pagina1_cotacao_02['indice05'].str.replace('I Preço Total\r','')
preco_total1 = preco_total1[preco_total1.str.len() > 0]
preco_total1 = preco_total1[~preco_total1.str.contains('Qtde.\r30')]
preco_total1 = preco_total1.reset_index(drop=True)
preco_total1 = preco_total1.apply(lambda x: str(x).replace(".",""))
preco_total1 = preco_total1.apply(lambda x: str(x).replace(",","."))
preco_total1 = preco_total1.astype('float64')

#mesclando os dados atrávés de cada indice formando uma tabela

df_mesclar01= descricao_material1.to_frame().merge(preco_unitario1,left_index=True,right_index=True)
df_mesclar01= descricao_material1.to_frame().merge(preco_unitario1,left_index=True,right_index=True)
Tabela_01= df_mesclar01.merge(preco_total1,left_index=True,right_index=True)
Tabela_01['indice02'] = Tabela_01['indice05'] / Tabela_01['indice04']


#display
#display(descricao_material1)
#display(quantidade_1)
#display(preco_unitario1)
#display(preco_total1)
Tabela_01
```
